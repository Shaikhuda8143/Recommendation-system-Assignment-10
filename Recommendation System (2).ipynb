# 4. Data Preparation

## User Based Similarity:

```python
books_data_df= pd.pivot_table( data=books_data,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
books_data_df
```

## Using Cosine as a metric 

```python
cosine_df = pd.DataFrame(1 - pairwise_distances(X = books_data_df,metric='cosine')).round(2)
cosine_df
```

## Applying Data Filtering

### Two methods of creating a recommendation systems:
**Content-based filtering** uses the description of the product or service and then recommends a product to that user to which his content matches.
**collaborative filtering** filters a group of people with similar characteristics to recommend products and services.

## Collaborative Filtering:
   >It recommends according to the user similarity, choices, mentality etc.
* This method is used to find correlations between different users and then recommend them the same products based on their similarities.
